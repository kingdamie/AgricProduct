### 1. Introduction
This notebook analyzes agricultural data from multiple countries to build predictive models for key agricultural metrics. The dataset contains information about crops, production, weather conditions, prices, and other relevant factors.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
import joblib
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Data Loading and Initial Exploration
First, let's load the dataset and perform initial exploration to understand its structure and contents.

```python
# Load the dataset
df = pd.read_csv('data/enhanced_agricultural_data.csv')

# Display basic information
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

# Check data types and missing values
print("\nData Types and Missing Values:")
display(df.info())

# Basic statistics for numerical columns
print("\nDescriptive Statistics:")
display(df.describe(include='all'))
```

Observations from Initial Exploration:
    The dataset contains [X] rows and [Y] columns

    Key numerical features: Area_harvested_ha, Production_tonnes, Yield_kg_per_ha, etc.

    Categorical features: Country, Crop, Policy_Flag

    Some missing values in certain columns that we'll need to handle

## 3. Data Preprocessing
Let's clean and prepare the data for modeling.

```python
# Handle missing values
# For numerical columns, fill with median (less sensitive to outliers)
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# For categorical columns, fill with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Verify no missing values remain
print("Missing values after imputation:")
print(df.isnull().sum())

# Convert Year to categorical (if appropriate)
df['Year'] = df['Year'].astype('object')

# Check for duplicates
print("\nNumber of duplicates:", df.duplicated().sum())

# Handle outliers (using IQR method)
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] > upper_bound, upper_bound, 
                         np.where(df[column] < lower_bound, lower_bound, df[column]))
    return df

# Apply to numerical columns (excluding Year if converted to categorical)
numerical_cols_for_outliers = [col for col in num_cols if col != 'Year']
for col in numerical_cols_for_outliers:
    df = handle_outliers(df, col)
```

## 4. Exploratory Data Analysis (EDA)
Let's explore the data to understand relationships and patterns.

```python
# Distribution of key numerical variables
num_cols_of_interest = ['Production_tonnes', 'Yield_kg_per_ha', 'Rainfall_mm', 'Temperature_C']
for col in num_cols_of_interest:
    plt.figure(figsize=(10, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Top 10 crops by average production
top_crops = df.groupby('Crop')['Production_tonnes'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_crops.values, y=top_crops.index, palette='viridis')
plt.title('Top 10 Crops by Average Production (tonnes)')
plt.xlabel('Average Production (tonnes)')
plt.ylabel('Crop')
plt.show()

# Production trends over years
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Production_tonnes', hue='Country', estimator='mean', ci=None)
plt.title('Production Trends Over Years by Country')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Correlation matrix
plt.figure(figsize=(12, 8))
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## 5. Feature Engineering
Let's create new features that might improve our model's performance.

```python
# Create yield per unit area (already exists, but we can create additional productivity metrics)
df['Productivity_index'] = df['Production_tonnes'] / df['Area_harvested_ha']

# Create climate interaction feature
df['Rainfall_Temp_interaction'] = df['Rainfall_mm'] * df['Temperature_C']

# Create price to yield ratio
df['Price_to_Yield_ratio'] = df['Price_USD_per_tonne'] / df['Yield_kg_per_ha']

# Create demand-supply balance feature
df['Demand_Supply_balance'] = np.where(df['Demand_Supply_Gap'] > 0, 'Surplus', 'Deficit')

# Log transform for skewed numerical features
skewed_cols = ['Production_tonnes', 'Area_harvested_ha', 'Transport_Cost_USD']
for col in skewed_cols:
    df[f'log_{col}'] = np.log1p(df[col])

# Display new features
print("New Features Created:")
print(df[['Productivity_index', 'Rainfall_Temp_interaction', 
          'Price_to_Yield_ratio', 'Demand_Supply_balance']].head())
```

## 6. Model Building
Now, let's build predictive models. We'll focus on predicting Yield (kg/ha) as our target variable.

```python


# 1. Data Preparation - Handle target first
df = df.dropna(subset=['Yield_kg_per_ha'])  # Drop rows with missing target
y = df['Yield_kg_per_ha']
X = df.drop(['Yield_kg_per_ha'], axis=1)

# 2. Clean numerical data
numerical_cols = X.select_dtypes(include=['float64', 'int64']).columns

# Handle infinite values and NaNs
X[numerical_cols] = X[numerical_cols].replace([np.inf, -np.inf], np.nan)
X[numerical_cols] = X[numerical_cols].fillna(X[numerical_cols].median())

# Cap extreme values
def cap_values(series, lower_percentile=0.01, upper_percentile=0.99):
    lower = series.quantile(lower_percentile)
    upper = series.quantile(upper_percentile)
    return series.clip(lower, upper)

for col in numerical_cols:
    X[col] = cap_values(X[col])

# 3. Handle categorical data
categorical_cols = X.select_dtypes(include=['object', 'category']).columns

# Convert all categorical columns to strings to avoid mixed types
for col in categorical_cols:
    X[col] = X[col].astype(str)

# 4. Create preprocessing pipelines
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)],
    remainder='drop')  # Explicitly drop any remaining columns

# 5. Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

# 6. Define and evaluate models
models = {
    'Random Forest': RandomForestRegressor(random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    'Linear Regression': LinearRegression()
}

results = {}
for name, model in models.items():
    try:
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('model', model)])
        
        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_test)
        
        results[name] = {
            'Model': pipeline,
            'MSE': mean_squared_error(y_test, y_pred),
            'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
            'R2': r2_score(y_test, y_pred)
        }
        
        print(f"{name}:")
        print(f"  MSE: {results[name]['MSE']:.2f}")
        print(f"  RMSE: {results[name]['RMSE']:.2f}")
        print(f"  R2: {results[name]['R2']:.2f}")
        print()
        
    except Exception as e:
        print(f"Error with {name}: {str(e)}")
        # For debugging, print the first few rows of problematic data
        print("Sample problematic data:")
        print(X_train.head())
        continue

# 7. Results analysis
if not results:
    print("\nNo models trained successfully. Checking data:")
    print("Numerical columns:", numerical_cols)
    print("Categorical columns:", categorical_cols)
    print("Missing values in X_train:", X_train.isna().sum().sum())
    print("Infinite values in X_train:", np.isinf(X_train[numerical_cols]).sum().sum())
    print("Missing values in y_train:", y_train.isna().sum())
else:
    best_model_name = max(results.items(), key=lambda x: x[1]['R2'])[0]
    print(f"\nBest model: {best_model_name} (R2: {results[best_model_name]['R2']:.2f})")
    
    # Get feature names for interpretation
    try:
        onehot_columns = list(
            results[best_model_name]['Model'].named_steps['preprocessor']
            .named_transformers_['cat']
            .named_steps['onehot']
            .get_feature_names_out(categorical_cols)
        )
        all_features = list(numerical_cols) + onehot_columns
        print(f"\nTotal features after preprocessing: {len(all_features)}")
    except Exception as e:
        print(f"Could not retrieve feature names: {str(e)}")
```

## 7. Model Evaluation
Let's compare the performance of our models and select the best one.

```python
# Create a comparison DataFrame
comparison_df = pd.DataFrame.from_dict(results, orient='index')[['MSE', 'RMSE', 'R2']]
display(comparison_df.sort_values('R2', ascending=False))

# Visualize model performance
plt.figure(figsize=(12, 6))
sns.barplot(x=comparison_df.index, y='R2', data=comparison_df.reset_index(), palette='viridis')
plt.title('Model Comparison by R-squared Score')
plt.ylim(0, 1)
plt.ylabel('R-squared')
plt.show()

# Select the best model based on R2 score
best_model_name = comparison_df['R2'].idxmax()
best_model = results[best_model_name]['Model']
print(f"\nBest Model: {best_model_name}")

# Plot actual vs predicted values for the best model
y_pred_best = best_model.predict(X_test)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred_best)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title(f'Actual vs Predicted Values ({best_model_name})')
plt.show()
```

## 8. Feature Importance Analysis
Understanding which features are most important in our best model.

```python
# Get feature names after preprocessing
# For OneHotEncoded features, we need to get the feature names
preprocessor.fit(X)
onehot_columns = list(preprocessor.named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features))
all_features = list(numerical_features) + onehot_columns

# Get feature importances from the best model
if hasattr(best_model.named_steps['model'], 'feature_importances_'):
    importances = best_model.named_steps['model'].feature_importances_
    feature_importance = pd.DataFrame({'Feature': all_features, 'Importance': importances})
    feature_importance = feature_importance.sort_values('Importance', ascending=False).head(20)
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
    plt.title('Top 20 Feature Importances')
    plt.show()
elif hasattr(best_model.named_steps['model'], 'coef_'):
    coefficients = best_model.named_steps['model'].coef_
    feature_importance = pd.DataFrame({'Feature': all_features, 'Coefficient': coefficients})
    feature_importance['Absolute_Coefficient'] = np.abs(feature_importance['Coefficient'])
    feature_importance = feature_importance.sort_values('Absolute_Coefficient', ascending=False).head(20)
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Coefficient', y='Feature', data=feature_importance, palette='viridis')
    plt.title('Top 20 Feature Coefficients')
    plt.show()
else:
    print("Feature importance not available for this model type.")
```
